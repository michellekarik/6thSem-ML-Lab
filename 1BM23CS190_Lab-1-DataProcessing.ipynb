```python
import pandas as pd
import numpy as np

df = pd.read_csv("adult.csv")
df
```

```python
df.replace('?', np.nan, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.info()
```

```python
num_colmns = df.select_dtypes(include=[np.int64]).columns
df[num_colmns] = df[num_colmns].fillna(df[num_colmns].mean())
df
```

```python
'''from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()   #create an instance of this object
for col in df.columns:     #for each column in columns
  if df[col].dtype =='object':  #categorical column data type == object
    df[col] = le.fit_transform(df[col])
df    '''
nominal_cols = df.select_dtypes(include=['object']).columns  #one hot encoding for nominal columns that have no meaning or order like male/female, mumbai/delhi/chennai
df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)
df
bool_cols = df.select_dtypes(include='bool').columns
df[bool_cols] = df[bool_cols].astype(int)
df
```

```python
# 1. Select only continuous numeric features
num_columns = df.select_dtypes(include=['int64', 'float64']).columns

# 2. Remove target column if present
num_columns = num_columns.drop('income_>50K', errors='ignore')

# 3. Remove binary (dummy) columns
num_columns = [col for col in num_columns if df[col].nunique() > 2]

Q1 = df[num_columns].quantile(0.25)
Q3 = df[num_columns].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (df[num_columns] < lower_bound) | (df[num_columns] > upper_bound)
df = df[~outliers.any(axis=1)]

df
```
